## Manejo de Fechas y Tiempos en Pandas

Pandas proporciona la funci√≥n `pd.to_datetime()` para convertir diferentes formatos de fechas en objetos `datetime`. Esto es √∫til para estandarizar fechas que pueden venir en distintos formatos y facilitar su manipulaci√≥n.

### Ejemplo 1: Convertir una columna de fechas

```python
import pandas as pd

# Crear un DataFrame con diferentes formatos de fecha
data = {'fechas': ['2025-02-11', '11/02/2025', 'Feb 11, 2025', '2025-02-11 14:30:00']}
df = pd.DataFrame(data)

# Convertir a datetime
df['fechas'] = pd.to_datetime(df['fechas'])
print(df)
```

### Opciones avanzadas de `pd.to_datetime()`
- `format='...'`: Para especificar un formato exacto.
- `dayfirst=True`: Indica que el primer valor es el d√≠a y no el mes.
- `errors='coerce'`: Convierte valores inv√°lidos en `NaT` en lugar de generar errores.

### Ejemplo 2: Manejo de fechas con formatos espec√≠ficos

```python
# Convertir fechas con un formato espec√≠fico
df['fechas'] = pd.to_datetime(df['fechas'], format='%Y-%m-%d', errors='coerce')
print(df)
```

Esto permite trabajar con fechas de manera flexible y efectiva.

## Formatos de Salida para Fechas y Tiempos en Pandas

Despu√©s de convertir una columna a `datetime`, puedes formatear su salida con `.dt.strftime('FORMATO_DESEADO')`.

### Ejemplo de Uso

```python
import pandas as pd

# Crear un DataFrame con una fecha en formato datetime
df = pd.DataFrame({'fechas': pd.to_datetime(['2025-02-11 14:30:45'])})

# Aplicar diferentes formatos
df['formato_1'] = df['fechas'].dt.strftime('%Y-%m-%d')
df['formato_2'] = df['fechas'].dt.strftime('%d/%m/%Y %H:%M:%S')
df['formato_3'] = df['fechas'].dt.strftime('%A, %d %B %Y')

print(df)
```

### Listado de formatos √∫tiles:
| **C√≥digo** | **Salida (Ejemplo: 2025-02-11 14:30:45)** | **Descripci√≥n** |
|------------|----------------------------------|----------------|
| `%Y-%m-%d` | `2025-02-11`  | A√±o-Mes-D√≠a (ISO 8601) |
| `%d-%m-%Y` | `11-02-2025`  | D√≠a-Mes-A√±o |
| `%m/%d/%Y` | `02/11/2025`  | Mes/D√≠a/A√±o (EEUU) |
| `%B %d, %Y` | `February 11, 2025`  | Nombre completo del mes |
| `%b %d, %Y` | `Feb 11, 2025` | Nombre corto del mes |
| `%A, %d %B %Y` | `Tuesday, 11 February 2025` | D√≠a de la semana con mes completo |
| `%a, %d %b %Y` | `Tue, 11 Feb 2025` | D√≠a corto de la semana con mes corto |
| `%Y/%m/%d %H:%M:%S` | `2025/02/11 14:30:45` | Fecha y hora completa (24h) |
| `%d-%m-%Y %I:%M %p` | `11-02-2025 02:30 PM` | Formato 12h con AM/PM |
| `%H:%M:%S` | `14:30:45` | Solo hora en formato 24h |
| `%I:%M:%S %p` | `02:30:45 PM` | Solo hora en formato 12h |
| `%j` | `042` | D√≠a del a√±o (1-365) |
| `%U` | `06` | N√∫mero de semana del a√±o (domingo como primer d√≠a) |
| `%W` | `06` | N√∫mero de semana del a√±o (lunes como primer d√≠a) |

Con esta tabla, puedes elegir el formato de salida que m√°s te convenga seg√∫n el contexto. üöÄ

## 2. Expresiones Regulares en Python


**Principales m√©todos de `re` (m√≥dulo de expresiones regulares):**

- `re.search(patron, texto)` ‚Üí Busca la primera coincidencia en el texto
- `re.findall(patron, texto)` ‚Üí Encuentra todas las coincidencias en el texto
- `re.match(patron, texto)` ‚Üí Verifica si el texto comienza con el patr√≥n
- `re.sub(patron, reemplazo, texto)` ‚Üí Reemplaza coincidencias en el texto

**Ejemplo de c√≥digo:**

```python
import re

texto = "Mi n√∫mero de tel√©fono es 654-123-456 y el de mi amigo es 987-654-321."
patron = r"\d{3}-\d{3}-\d{3}"

# Encontrar todos los n√∫meros de tel√©fono
telefonos = re.findall(patron, texto)
print(telefonos)
```

## Expresiones Regulares en Python (Ampliado)

Las expresiones regulares (regex) son patrones utilizados para buscar, extraer o manipular texto.

### Principales M√©todos de `re`
| M√©todo | Descripci√≥n |
|--------|------------|
| `re.search(patron, texto)` | Busca la **primera** coincidencia en el texto |
| `re.findall(patron, texto)` | Devuelve **todas** las coincidencias como lista |
| `re.match(patron, texto)` | Comprueba si el **inicio** del texto coincide con el patr√≥n |
| `re.sub(patron, reemplazo, texto)` | Reemplaza coincidencias en el texto |
| `re.split(patron, texto)` | Divide el texto seg√∫n el patr√≥n |

### Caracteres Especiales en Regex
| Patr√≥n | Significado | Ejemplo | Resultado |
|--------|------------|---------|-----------|
| `\d` | Un d√≠gito (0-9) | `re.findall(r"\d", "A√±o 2024")` | `['2', '0', '2', '4']` |
| `\D` | No es un d√≠gito | `re.findall(r"\D", "A√±o 2024")` | `['A', '√±', 'o', ' ']` |
| `\w` | Caracter alfanum√©rico (A-Z, a-z, 0-9, _) | `re.findall(r"\w", "Python_3!")` | `['P', 'y', 't', 'h', 'o', 'n', '_', '3']` |
| `\W` | No es alfanum√©rico | `re.findall(r"\W", "Python_3!")` | `['!']` |
| `\s` | Espacio en blanco | `re.findall(r"\s", "Hola Mundo")` | `[' ']` |
| `\S` | No es un espacio en blanco | `re.findall(r"\S", "Hola Mundo")` | `['H', 'o', 'l', 'a', 'M', 'u', 'n', 'd', 'o']` |
| `.` | Cualquier caracter excepto salto de l√≠nea | `re.findall(r"a.b", "axb acb a9b")` | `['acb', 'a9b']` |
| `^` | Inicio de l√≠nea | `re.search(r"^Hola", "Hola mundo")` | `Match encontrado` |
| `$` | Fin de l√≠nea | `re.search(r"mundo$", "Hola mundo")` | `Match encontrado` |
| `[]` | Conjunto de caracteres permitidos | `re.findall(r"[aeiou]", "Python")` | `['o']` |
| `[^ ]` | Niega el conjunto | `re.findall(r"[^aeiou]", "Python")` | `['P', 'y', 't', 'h', 'n']` |
| `*` | 0 o m√°s repeticiones | `re.findall(r"ab*", "ab abbb abc")` | `['ab', 'abbb', 'ab']` |
| `+` | 1 o m√°s repeticiones | `re.findall(r"ab+", "ab abbb abc")` | `['ab', 'abbb']` |
| `?` | 0 o 1 repetici√≥n | `re.findall(r"ab?", "ab ac abc")` | `['ab', 'a', 'ab']` |
| `{n}` | Exactamente n repeticiones | `re.findall(r"\d{3}", "1234 567 89")` | `['123', '567']` |
| `{n,}` | Al menos n repeticiones | `re.findall(r"\d{2,}", "1 12 123 1234")` | `['12', '123', '1234']` |
| `{n,m}` | Entre n y m repeticiones | `re.findall(r"\d{2,4}", "1 12 123 1234 12345")` | `['12', '123', '1234']` |
| `|` | Alternativa (o) | `re.findall(r"Hola|Adi√≥s", "Hola mundo. Adi√≥s amigo.")` | `['Hola', 'Adi√≥s']` |
| `()` | Agrupar expresiones | `re.search(r"(Hola) (Mundo)", "Hola Mundo")` | `Match con grupos` |

## Uso de APIs en Python (Ampliado)

Las **APIs** (Application Programming Interfaces) permiten que aplicaciones se comuniquen entre s√≠, intercambiando datos.

### ¬øC√≥mo funciona una API?

1. Un cliente (nuestro c√≥digo) hace una **solicitud HTTP** a un servidor.
2. El servidor procesa la solicitud y responde con **datos en formato JSON, XML u otro**.
3. El cliente recibe los datos y los procesa seg√∫n sea necesario.

### M√©todos HTTP M√°s Usados
| M√©todo | Descripci√≥n |
|--------|------------|
| `GET`  | Solicitar informaci√≥n (ej. obtener datos) |
| `POST` | Enviar datos al servidor |
| `PUT`  | Modificar datos en el servidor |
| `DELETE` | Eliminar datos en el servidor |

### M√≥dulo `requests` en Python

Python tiene la librer√≠a `requests`, que permite interactuar con APIs de forma sencilla.

#### 1. Hacer una Solicitud GET
```python
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"  # URL de prueba

response = requests.get(url)  # Hacer la petici√≥n GET

if response.status_code == 200:  # Verificar si la respuesta es exitosa
    data = response.json()  # Convertir la respuesta a JSON
    print(data)
else:
    print("Error en la solicitud:", response.status_code)
```

#### 2. Hacer una Solicitud POST (Enviar Datos)
```python
url = "https://jsonplaceholder.typicode.com/posts"

nuevo_post = {
    "title": "Mi nuevo post",
    "body": "Este es el contenido del post",
    "userId": 1
}

response = requests.post(url, json=nuevo_post)

if response.status_code == 201:
    print("Post creado con √©xito:", response.json())
else:
    print("Error al crear el post")
```

#### 3. Extraer Informaci√≥n de una API Real (Ejemplo con OpenWeather)
```python
API_KEY = "tu_api_key"  # Reemplaza con una clave v√°lida

```

### Consejos para Trabajar con APIs

1. **Leer la documentaci√≥n**: Cada API tiene su propia estructura y requisitos.
2. **Manejar errores**: Usar `response.status_code` para verificar respuestas exitosas.
3. **Autenticaci√≥n**: Algunas APIs requieren claves de acceso o tokens (`headers = {'Authorization': 'Bearer TU_TOKEN'}`).
